```python
def CreateReamingPaths(Source,Destination):
    ans=[]
    print(Source,Destination)
    name1 = 'ipxls'
    name2 = '.xls'
    path = 'C:\ip4d'
    while(Source<Destination-1):
        print('S++ = ',Source+1)
        NAME = name1+str(Source+1)+name2
        print(NAME)
        ans.append(os.path.join(path,NAME))
        print(os.path.join(path,NAME))
        Source = Source +1
    return ans
```
